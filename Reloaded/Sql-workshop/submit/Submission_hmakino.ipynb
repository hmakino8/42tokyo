## 課題1: Popular Recommend

劇場アニメとTV&OVAアニメそれぞれについて、人気作品を上位50件ずつ取得し、個別に表示したリコメンド。

```python
%%sql
/*
animeテーブルのtype 'TV'と'OVA'を'TV&OVA'に書き換えたテーブルanime2を作成。
*/

WITH anime2 AS (
    SELECT
        *
        , CASE type
            WHEN 'TV' THEN
                'TV&OVA'
            WHEN 'OVA' THEN
                'TV&OVA'
            ELSE
                type
        END AS media
    FROM anime
)

/*
上記で作成したanime2テーブルをもとに、type'Movie'と'TV&OVA'別にtop50を表示
*/

SELECT
    *
FROM ( 
   SELECT
        media
        , rank () OVER (PARTITION BY media ORDER BY ranked) AS rank
        , japanese_name AS "劇場アニメ、TV&OVAアニメ別ランキング　TOP50！"
    FROM anime2
 ) AS tmp
WHERE
    rank <= 50 AND media IN ('Movie', 'TV&OVA')
```

## 課題2: Item to Item Recommend

作成したレコメンドの説明

特定のアニメについて、animeテーブルの'ranked'レコードtop100かつジャンルマッチ数の高い他の作品top20をリコメンド

```python
%%sql
/*
    特定のアニメと他のアニメのジャンルごとのマッチ判定(true or false)
    を記載したテーブルを作成
*/

WITH genre_match AS (
    SELECT
        ranked
        , mal_id
        , japanese_name
        , string_to_array(unnest(string_to_array(genres,',')),',') && (
            SELECT string_to_array(genres,',')
            FROM anime
            WHERE mal_id = 123) AS match
    FROM anime
)

/*
    特定のアニメと他のアニメのジャンルごとのマッチ判定(1 or 0)
    を記載したテーブルを作成
*/

, match1_0 AS (
    SELECT
        ranked
        , mal_id
        , japanese_name
        , CASE match
            WHEN 'True' 
                THEN 1
            ELSE
                0
            END AS is_true
    FROM genre_match
)

/*
    特定のアニメと他のアニメのジャンルごとのマッチ数
    を記載したテーブルを作成
*/

, match_count AS (
    SELECT
        ranked
        , mal_id
        , japanese_name
        , SUM(is_true) AS count
    FROM match1_0
    GROUP BY mal_id, japanese_name, ranked
)

/*
    上記テーブル(match_count)をもとに
    animeテーブルの'ranked'レコードtop100かつジャンルマッチ数top20を表示
*/

SELECT
    count AS ジャンルマッチ数
    , japanese_name AS このアニメに似たジャンルのアニメ　20選！
FROM match_count
WHERE ranked <= 100
ORDER BY count DESC, ranked
LIMIT 20
```

## 課題3: User to Item Recommend

作成したレコメンドの説明

animeテーブルの'ranked'レコードのうち、 特定ユーザーのwatching_statusが1(現在視聴中)、2(視聴完了)、4(視聴をやめた)のanimeを除いたtop10をリコメンド。

```python
%%sql
/*
    特定ユーザーのwatching_statusが3(保留), 6(視聴予定)以外のステータスの
    アニメのリストを作成
*/

WITH animelist_for_each_user AS (
    SELECT 
        *
    FROM 
        animelist
    WHERE 
        user_id = 12345678 
        AND watching_status NOT IN(3, 6)
)

/*
    animeテーブルに上記テーブルを結合し、テーブルの'ranked'レコードのうち、
    上記テーブルに記載されたアニメを除いたtop10を表示
*/

SELECT
    a.japanese_name AS "あなたにおすすめのアニメ　10選！"
FROM anime AS a
    FULL JOIN animelist_for_each_user AS u
    ON a.mal_id = u.anime_id
GROUP BY
    a.mal_id,
    a.japanese_name,
    a.ranked
HAVING count(u.user_id) = 0 AND a.ranked > 0
ORDER BY a.ranked
LIMIT 10;
```
